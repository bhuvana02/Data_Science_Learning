### Q 7: Generate additional feature named F_X such that :-
* International plan is True then the value of F_X is 1
* If International plan is False and Account Length is more than the average Account length then F_X is 1
* If Number vmail messages is greater than 75th percentile and Customer Service calls is greater than 1 then F_X is 1
* In all other cases F_X is 0

```python
data.columns
```

```python
data['F_X']=0
for i,j in enumerate(data['F_X']):
    v=data['Number vmail messages'][i]>data['Number vmail messages'].quantile(0.75)
    w=data['Customer service calls'][i]>1
    if data['International plan'][i]==True:
        data['F_X'][i]=1
    elif data['Account length'][i]>data['Account length'].mean():
        data['F_X'][i]=1
    elif v==True and w==True:
        data['F_X'][i]=1  
```

### 4.pipe()

pipe function organizes multiple preprocessing operations into a single one

```python
def fill_missing_values(df):
   for col in df.select_dtypes(include= ["int","float"]).columns:
      val = df[col].mean()
      df[col].fillna(val, inplace=True)
   return df
def remove_outliers(df, column_list):
   for col in column_list:
      avg = df[col].mean()
      std = df[col].std()
      low = avg - 2 * std
      high = avg + 2 * std
      df = df[df[col].between(low, high, inclusive=True)]
   return df
```

```python
data1 = (data.pipe(fill_missing_values).
                      pipe(remove_outliers, ['Total day minutes', 'Total day charge', 'Total eve minutes',
       'Total eve charge', 'Total night minutes', 'Total night charge',
       'Total intl minutes', 'Total intl charge']))
```

### 10.argpartition()

argpartition() is use to find N largest values index.
The output will be the N largest values index, 
and then we can sort the values if needed.

```python
x = np.array([12, 10, 12, 0, 6, 8, 9, 1, 16, 4, 6, 0])
index_val = np.argpartition(x, 5)[5:]
index_val
```

```python
np.sort(x[index_val])
```
